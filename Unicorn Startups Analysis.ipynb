# Dropping the NaN values

```python
data.dropna(inplace=True)
data
```

```python
data.isna().sum()
```

```python
data['Valuation ($B)']=data['Valuation ($B)'].str.replace('$', '',regex=True)
data['Valuation ($B)']=data['Valuation ($B)'].astype(float)
data
```

# Visualisation

```python
plt.figure(figsize=(10,10),dpi=200)
sns.countplot(y=data.Country)
plt.show()

# Total Startups produced by each country so far till 2022
```

```python
plt.figure(figsize=(10,10),dpi=200)
sns.countplot(y=data.Industry)
plt.show()

# Total no.of industry types produced till 2022 in the world
```

```python
data.groupby('Country')['Valuation ($B)'].sum().sort_values(ascending=True).plot.barh(figsize=(20,20))
plt.show()

# The Total Valuations seeked by all the companies belonging to a particular nation
```

# Top 10 Unicorns in the world by Valuation

```python
top_companies = data.sort_values(by = 'Valuation ($B)', ascending=False).head(10)
plt.figure(figsize=(15,15),dpi=200)
fig = sns.barplot(data = top_companies, x = 'Company', y = 'Valuation ($B)',hue = 'Industry', dodge=False, palette = 'rocket')
ticks = fig.set_xticklabels(fig.get_xticklabels())
labels = fig.set(xlabel ="Company", ylabel = "Valuation ($B)", title ='Top 10 companies by valuation')
legend = plt.legend(title='Industry', loc='upper right')
```

# Top 10 Indian Unicorns By Valuation

```python
top10indian = data[data.Country=='India'].head(71)
top10indian
```

```python
top_companies = top10indian.sort_values(by = 'Valuation ($B)', ascending=False).head(10)
plt.figure(figsize=(30,28),dpi=200)
fig = sns.barplot(data = top_companies, x = 'Company', y = 'Valuation ($B)',
                  hue = 'Industry', dodge=False, palette = 'seismic_r',orient = "v" )
ticks = fig.set_xticklabels(fig.get_xticklabels())
labels = fig.set(xlabel ="Company", ylabel = "Valuation ($B)", title ='Top 10 companies of India by valuation')
legend = plt.legend(title='Industry', loc='upper right')
```

Important thing to notice here is that India was more succesful in terms of launching more number of unicorn startups than the UK but when the sum of valuations of all the companies from a particular country are being taken into account UK surpasses India only by a small margin which means that even though there might be less unicorn turned startups in UK than India they are valued higher
